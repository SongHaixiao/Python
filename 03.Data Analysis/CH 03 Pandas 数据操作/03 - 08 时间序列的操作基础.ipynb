# date_range() : 产生区间数据

- 参数：

    - start : 开始 index;
    - end : 结束 index;
    - periods :数据容量，即 产生的数据数量；
    - freq 计算单位 ，参数：
        - D - day 天 : 以天为单位； 
        - W - week 周 : 以周为单位;  
        - W - MON 周一 ： 以周一为单位;
        - nH 小时 : 以小时为单位, H 为 1小时，n为数字，即 nH 为 n小时。

```python
# 生成 随机 date 时间
# 起始时间：2016-01-01
# 数据容量 : 100 个
# 计算间隔 : 按天算
date_list_new = pd.date_range('2016-01-01',periods=100)
date_list_new
```

```python
# 生成 随机 date 时间
# 起始时间：2016-01-01
# 数据容量 : 100 个
# 计算间隔 : 按周算
date_list_new = pd.date_range('2016-01-01',periods=100,freq='W')
date_list_new
```

```python
# 生成 随机 date 时间
# 起始时间：2016-01-01
# 数据容量 : 100 个
# 计算间隔 : 一个小时
date_list_new = pd.date_range('2016-01-01',periods=100,freq='H')
date_list_new
```

```python
# 生成 随机 date 时间
# 起始时间：2016-01-01
# 数据容量 : 100 个
# 计算间隔 : 小时
date_list_new = pd.date_range('2016-01-01',periods=100,freq='H')
date_list_new
```

```python
# 生成 随机 date 时间
# 起始时间：2016-01-01
# 数据容量 : 100 个
# 计算间隔 : 5小时
date_list_new = pd.date_range('2016-01-01',periods=100,freq='5H')
date_list_new
```

```python
s2 = Series(np.random.rand(100),index = date_list_new)
s2
```
