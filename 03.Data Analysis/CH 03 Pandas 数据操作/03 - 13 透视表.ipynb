```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# 读取 sales-funnel.xlsx 文件
df = pd.read_excel('sales-funnel.xlsx')
df
```

# pivot_table() ：透视表函数,使用透视表的前提是对数据之间的关系有大体地了解

- 参数：

    - first : DataFrame Type
    - index : 基于某些列建立透视表,这些列之间有一定的关系；
    - aggfunc : 默认为求平均数函数；可以改为其他函数：sum等；
    - values : 求特定列的值;
    - columns : 在 values 下面 再细分一些 columns;
   

```python
pd.pivot_table(df,index=['Name']) # 对原表数据求平均
```

```python
pd.pivot_table(df, index = ['Name'],aggfunc = 'sum') # 以 Name 列为 index,对原表数据求平均
```

```python
pd.pivot_table(df, index = ['Name','Rep','Manager']) # 以 Name,Rep,Manager 为index,对原表数据求平均
```

```python
# 以 Manager、Rep 为 index，对 Price 求平均值
# 其中，一个 Manager 下 有好几个 Rep 销售代表
pd.pivot_table(df, index = ['Manager','Rep'],values=['Price']) 
```

```python
# 以 Manager、Rep 为 index，对 Price 求和
# 其中，一个 Manager 下 有好几个 Rep 销售代表
pd.pivot_table(df, index = ['Manager','Rep'],values=['Price'],aggfunc='sum') 
```

```python
# 以 Manager、Rep 为 index，对 Price 和 Quantity 求和
# 其中，一个 Manager 下 有好几个 Rep 销售代表
pd.pivot_table(df, index = ['Manager','Rep'],values=['Price','Quantity'],aggfunc='sum') 
```

```python
# 以 Manager、Rep 为 index，对 Price 和 Quantity 下的 Product 各列求平均值
# 对原来值为 NaN 的情况，填充 0
pd.pivot_table(df, index = ['Manager','Rep'],values=['Price','Quantity'],columns='Product',fill_value=0) 
```
