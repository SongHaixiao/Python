## apply (function)
- function : 对单个元素进行操作的元素

## Question

data 值每一行的中的 Symbol、Seqno、Price 冒号后的值

```python
# 得到单行值的操作
# 将 index 为 0 的 data 值 去掉空格切片
l1 = df['data'][0].strip().split(' ')
l1[1],l1[3],l1[5]
```

```python
# 得到多行值的操作
# 定义一个函数
def foo(line):
    items = line.strip().split(' ')
    return Series([items[1],items[3],items[5]])
```

```python
df_tmp = df['data'].apply(foo)
df_tmp
```

```python
# Rename 3 个 列值的名字
df_tmp = df_tmp.rename(columns={0:"Symbol",1:"Seqno",2:"Price"})
df_tmp
```

```python
# 将 df_tmp 与 df 进行 combine
df_tmp
```

```python
df
```

```python
df_new = df.combine_first(df_tmp)
df_new
```

```python
# 删除 data column
del df_new['data']
df_new
```

```python
# 删除 A column
del df_new['A']
df_new
```

```python
# 到处 csv 文件
df_new.to_csv('demo_duplicate.csv')
```
