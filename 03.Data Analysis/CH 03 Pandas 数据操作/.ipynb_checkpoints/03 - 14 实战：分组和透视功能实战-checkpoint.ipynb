# 1. 获取最长延误时间 Top 10

```python
# 对数据 按照延迟时间 进行降序排序
# 只看前 10 条
df.sort_values('arr_delay', ascending = False)[:10]
```

# 2. 计算延误和没有延误所占比例

- value_counts : 计算某一列中各取值的统计情况

### Task 1 : 查看 取消与否的航班数

```python
# 0 ：没有取消
# 1 ：取消
df['cancelled'].value_counts()
```

### Task 2 : 生成一个新的 delayed 列 来判断航班是否延迟

```python
# 操作：对 arr_delay 中数据进行
df['delayed'] = df['arr_delay'].apply(lambda x : x > 0)
```

```python
df.head()
```

### Task 3 : 查看延误比 

```python
# 计算 delayed 列中 无延误 Fals 和 延误 True 各自总数
delay_data = df['delayed'].value_counts()
```

```python
type(delay_data) # Series 数据类型
```

```python
# delay_data[1] 是延误的
# delay_data[0] 是没有延误的
delay_data[1]/(delay_data[0]+delay_data[1])
```

### Task 4 : 每一个航空公司延误的情况

```python
delay_group = df.groupby(['unique_carrier','delayed'])
```

```python
delay_group
```

```python
# 用  size() 返回 多级 groupby()
# 第一级 ： unique_carrier,delayed
# 第二级 ： AA , False or Ture
# 第三级 ： 数值
delay_group.size()
```

```python
# unstack() 将 size() 得到的数据 转换成 DataFrame 
delay_group.size().unstack()
```
