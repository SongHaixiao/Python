# pivot_table() ：透视表函数,求平均值

- 参数：

    - first : DataFrame Type
    - index : 基于某些列建立透视表
    - aggfunc : 默认为求平均数函数；可以改为其他函数：sum、
   

```python
pd.pivot_table(df,index=['Name']) # 对原表数据求平均
```

```python
pd.pivot_table(df, index = ['Name'],aggfunc = 'sum') # 对原表数据求平均
```
