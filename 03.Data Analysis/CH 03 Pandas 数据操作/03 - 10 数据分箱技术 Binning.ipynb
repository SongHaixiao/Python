```
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

Question：

将 分数 按照 (0,59]，(60,70]，(71,80]，(81,100] 的区间进行分割 

```
# create a random score list
score_list = np.random.randint(25,100,size=20)
```

```
score_list
```

```
bins = [0,59,70,80,100]
```

# cut() : 判断各数据属于哪些范围

- 参数：

  - first parameters : 数据对象；
  - second parameters : 分箱范围；
  - labels : 对各区间取别名。


```
score_cat = pd.cut(score_list,bins)
score_cat
```

```
score_cat = pd.cut(score_list,bins,labels = ['Low','OK','Good','Great'])
score_cat
```

# value_counts(): 计算各值出现的次数

```
pd.value_counts(score_cat)
```

```
df = DataFrame()
```
