# 作业描述

- 将 movie_metadata.csv 文件 导入并将所有数据按照以 imdb_score 数据降序 标准进行排序

- 并将生成的数据导出成新的 csv 文件imdb.csv

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
csv_input = 'movie_metadata.csv'
```

# read_csv() 读取 csv 文件

### head() 

- 无参数，返回所有列的前5行数据

```python
pd.read_csv(csv_input).head() # 默认无参数时，返回所有列的前5行数据
```

###  head()[["column1_name","column2_name",..]].head()

- 带有参数，查找指定列的前5行内容


```python
pd.read_csv(csv_input).head()[["movie_title","director_name","imdb_score"]].head()
```

### sort_values()

- 按指定列数据的值排序
- 第一个参数为指定列的名字，必须有
- ascending 参数， 默认为 True 升序排序；设置为 False 时，降序排序.

```python
pd.read_csv(csv_input).head()[["movie_title","director_name","imdb_score"]].head().sort_values("imdb_score")
```

## 无 head() 时， 获取指定列的所有数据内容

```python
# return 所有数据
pd.read_csv(csv_input)[["movie_title","director_name","imdb_score"]]
```

```python
# 给所有数据按照 imdb_scores 值升序排序
pd.read_csv(csv_input)[["movie_title","director_name","imdb_score"]].sort_values("imdb_score")
```

```python
# 给所有数据按照 imdb_scores 值降序排序
pd.read_csv(csv_input)[["movie_title","director_name","imdb_score"]].sort_values("imdb_score",ascending=False)
```

```python
# 将所有数据按照 imdb_scores 值降序排序的数据导出成 csv 文件
pd.read_csv(csv_input)[["movie_title","director_name","imdb_score"]].sort_values("imdb_score",ascending=False).to_csv("imdb.csv")
```
