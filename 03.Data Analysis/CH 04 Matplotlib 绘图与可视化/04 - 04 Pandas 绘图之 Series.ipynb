# plot():

- 参数 :

    - kind : 图形形状,默认为 line;
    - grid : 设置背景为 网格状，默认为 False - 没有网格效果，True - 设置背景为 网格状；
    - label : 图例名称;
    - title : 图标名称；
    - style : 线条格式，默认为实线，'--' 为虚线。

# plt.legend(): 显示图例名称

```python
# 单个例子
# 随机产生 1000 个随机小数求累加和
s1 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind='line',grid='True',label='S1',title='This is Series',style='--')
plt.legend()
```

```python
# 多个例子
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind='line',grid=True,label='S1',title='This is Series',style='--')
s2.plot(label='S2')
plt.legend()
```

# 绘制子图

### 直接通过 ax 调用 子图

```python
# 通过子图来画两个数据
fig, ax = plt.subplots(2,1)
```

```python
ax
```

```python
# 通过 ax 调用子图
ax[0].plot(s1)
```

```python
# 通过 ax 调用子图
ax[1].plot(s2)
```

### 通过 plot() 函数中，ax 参数来调用子图

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
fig, ax = plt.subplots(2,1)
```

```python
s1[0:10].plot(ax=ax[0],label='S1',kind='bar')
s2.plot(ax=ax[1],label='S2')
```
