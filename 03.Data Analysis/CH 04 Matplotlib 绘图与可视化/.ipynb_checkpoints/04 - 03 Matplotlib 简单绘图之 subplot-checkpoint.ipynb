```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0,5.0)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.plot(x,y1,'b--',label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x,y2,'r--',label='sin(pi*2x)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('This is x-y value')
plt.legend()
```

# subplot(): 绘制子图
- 将多图切分成独立的图
- 参数：
    - first : 规定行数；
    - second : 规定列数；
    - third : 选择画图的行数;
    - 前三个参数之间的逗号可以取消；
    - label ：标签名字。

```python
# 2 行 1 列，选择第一个位置
plt.subplot(2,1,1)
plt.plot(x,y1,'b--')
plt.ylabel('y1')
plt.subplot(2,1,2)
# 2 行 1 列，选择第二个位置
plt.plot(x,y2,'r--')
plt.ylabel('y2')
plt.xlabel('x')
```

```python
# 2 行 2 列，选择第一个位置
plt.subplot(221)
plt.plot(x,y1,'b--')
plt.ylabel('y1')

# 2 行 2 列，选择第二个位置
plt.subplot(222)
plt.plot(x,y2,'r--')
plt.ylabel('y2')
plt.xlabel('x')

# 2 行 2 列，选择第三个位置
plt.subplot(223)
plt.plot(x,y1,'r*')

# 2 行 2 列，选择第四个位置
plt.subplot(224)
plt.plot(x,y1,'b*')
```

## subplots() 返回两个对象

- 一个是 FigureSubplot 对象, 另一个是 AxesSubplot 对象
- 可以通过 AxesSubplot 对象来调用 plot() 方法将图画到 FigureSubplot 上

```python
# 无参数，单个图
figure, ax = plt.subplots()
```

```python
# 给图赋值
ax.plot([1,2,3,4,5])
```

```python
# 显示图
figure 
```

```python
# 多参数，2 行 2 列图
figure, ax = plt.subplots(2,2)
```

```python
# 查看 ax 数据类型
ax
```

# 例子

```python
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
```
