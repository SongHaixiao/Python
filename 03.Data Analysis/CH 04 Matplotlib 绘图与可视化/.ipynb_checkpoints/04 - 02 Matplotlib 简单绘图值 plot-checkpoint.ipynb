```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# plot():
- 单参数：
    - y 坐标 ：a 的取值
    - x 坐标 ：取值所对应的 index
    
- 双参数：
    - x 坐标：第一个参数的值
    - y 坐标： 第二个参数的值
    - 两个参数的长度必须一样，否则报错
    
- 三参数：
    - x 坐标：第一个参数的值
    - y 坐标: 第二个参数的值
    - 第三个参数：画图格式,前面英文字母为选定颜色
- 多参数：在同一个框中中显示不同的图片
    - 每两个参数为一组画图
    - 可选定各自画图格式
- label 变量：设置曲线名称

```python
a = [1,2,3]
```

```python
plt.plot(a)
```

```python
b = [4,5,6]
plt.plot(a,b)
```

```python
# %timeit 计算代码执行时间
%timeit np.arange(10)
```

```python
# a,b两参数，画出 以 * 为格式画图
plt.plot(a,b,'*')
```

```python
# a,b两参数，画出 以 红色的 -- 为格式画图
plt.plot(a,b,'r--')
```

```python
# 以 a,b,c,d 四个参数分别画两个图，a,b 为 红色--,c,d为 蓝色 *
c = [10,8,6]
d = [1,8,3]
plt.plot(a,b,'r--',c,d,'b*')
```

# 图片标识：
- xlabel() : x 轴 label；
- ylable() : y 轴 label；
- title() : 图片名称；
- legend() : 显示样例


```python
plt.plot(t,s,'r--',t*2,s,'--')
```

```python
plt.plot(t,s,'r--',label='aaa')
plt.plot(t*2,s,'b--',label='bbb')
plt.xlabel('This is x')
plt.ylabel('This is y')
plt.title('This is a demo')
plt.legend()
```
