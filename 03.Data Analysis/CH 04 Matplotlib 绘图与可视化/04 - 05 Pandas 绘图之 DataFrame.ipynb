```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
df = DataFrame(
     np.random.randint(1,10,40).reshape(10,4),
     columns = ['A','B','C','D']
)
```

```python
df
```

# 使用 plot() 方法对 DataFrame 进行绘图

## 参数

### kine : 图像格式

- line : 线性图；

- bar : 竖状柱状图；

- barh : 横状柱形图；

- area : 区域图；

### stacked : 同一个 index 内，将不同的数据块进行堆叠

- True ：真；
- False ：假，默认情况。

```python
# 竖状柱状图
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar',stacked=True)
```

```python
# 横状柱形图
df.plot(kind = 'barh')
```

```python
# 线性图
df.plot(kind = 'line')
```

```python
# 区域图
df.plot(kind='area')
```

# 对 DataFrame 中某一行绘图
### iloc[n]:
- 返回 DataFrame 中 第 n 行的取值
### plot()：
- 绘图

```python
# 对 A,B,C,D 第 5 行的取值
a = df.iloc[5]
a
```

```python
type(a)
```

```python
df.iloc[5].plot()
```

# 对每一行的数据绘图：

### 1. 通过 for 循环

```python
for i in df.index:
    df.iloc[i].plot(label=str(i))
plt.legend()
plt.show()
```

```python
df['A'].plot()
```
