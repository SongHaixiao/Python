```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# 用 matplotlib 画图

```python
def sinplot():
    x = np.linspace(0,14,100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x,np.sin(x+i)*(i+0.75),label='sin(x+%s)*(%s+0.75)'%(i,i))
    plt.legend()
```

```python
sinplot()
```

# Seaborn 调色

```python
import seaborn as sns
```

```python
sinplot()
```

# 调色板
## color_palette() :调色板函数

- 参数：
    - list 数据格式
    - 6 个 tuple
    - 每个 tuple 有三个值 为 RGB 值
    
## palplot(): 显示调色板的颜色
- 参数：

    - pal : 调色板函数
    - size : 色块儿个数
    
- 默认主题色板：
    - deep;
    - muted;
    - pastel;
    - bright;
    - dark;
    - colorblind.

```python
# 系统默认调色板
sns.color_palette()
```

```python
sns.palplot(sns.color_palette())
```

## set_palette() : 设置调色板
- 参数：

    - palette : 调色板函数；


```python
# 默认调色板样式
sinplot()
```

```python
# dark 调色板模式
sns.set_palette(sns.color_palette('dark'))
sinplot()
```

## set()：恢复到默认状态

```python
sns.set()
```

```python
sinplot()
```

## 定义任意长度的色板

```python
# 两个色板
pall = sns.color_palette([(0.5,0.1,0.7),(0.3,0.1,0.9)])
sns.palplot(pall)
```

```python
sns.palplot(sns.color_palette('hls',8))
```

## with 方法：临时更改系统画板

```python
# 临时将色板设置为 dark 主题
with sns.color_palette('dark'):
    sinplot()
```

```python
# 系统色板不变
sinplot()
```
