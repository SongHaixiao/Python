```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
%matplotlib inline
```

```python
import seaborn as sns
```

```python
s1 = Series(np.random.randn(1000))
```

# 通过 matplotlib 画直方图和密度图

```python
# 通过 matplotlib 画直方图
plt.hist(s1)
```

```python
# 通过 matplotlib 画密度图
s1.plot(kind='kde')
```

# 通过 seaborn 画 直方图和密度图

## distplot(): 绘制直方图与密度图函数
### 参数：
- data : 目标数据；
- bins : rug 区间个数；
- hist : 是否绘制直方图，默认为 False - 不绘制，True - 绘制；
- kde ：是否绘制密度图，默认为 False - 不绘制，True - 绘制；
- rug : 各区间数据分布情况，默认为 False - 不绘制，True - 绘制。

```python
sns.distplot(s1,hist=True,kde=True)
```

```python
sns.distplot(s1,hist=True,kde=True,rug=True)
```

```python
sns.distplot(s1,bins=20,hist=True,kde=False,rug=True)
```

# kdeplot()
## 参数：
- data : 目标数据；
- shade : 背景填充，True - 填充，False - 不填充；
- color : 颜色参数，‘r'- 红色；

```python
sns.kdeplot(s1,shade = True, color='r')
```
